### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento total da Loja 1: R$ {faturamento_loja1:,.2f}")
print(f"Faturamento total da Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento total da Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento total da Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
import matplotlib.pyplot as plt

faturamentos = {
    'Loja 1': faturamento_loja1,
    'Loja 2': faturamento_loja2,
    'Loja 3': faturamento_loja3,
    'Loja 4': faturamento_loja4
}

lojas = list(faturamentos.keys())
valores = list(faturamentos.values())

plt.figure(figsize=(8, 6))
plt.bar(lojas, valores, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Faturamento Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--')
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Combine sales by category from all stores into a single DataFrame
vendas_combinadas = pd.DataFrame({
    'Loja 1': vendas_por_categoria_loja,
    'Loja 2': vendas_por_categoria_loja2,
    'Loja 3': vendas_por_categoria_loja3,
    'Loja 4': vendas_por_categoria_loja4
}).reset_index()

vendas_combinadas_melted = vendas_combinadas.melt(id_vars='Categoria do Produto', var_name='Loja', value_name='Faturamento (R$)')

# Create the grouped bar chart
plt.figure(figsize=(15, 8))
sns.barplot(x='Categoria do Produto', y='Faturamento (R$)', hue='Loja', data=vendas_combinadas_melted, palette='tab10')
plt.title('Vendas por Categoria de Produto em Todas as Lojas')
plt.xlabel('Categoria do Produto')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Loja')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes = [loja['Frete'].mean(), loja2['Frete'].mean(), loja3['Frete'].mean(), loja4['Frete'].mean()]
lojas_fretes = pd.DataFrame({'Loja': lojas, 'Frete Médio': fretes})
lojas_fretes
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Calculate the mean freight for each store
mean_frete_loja1 = loja['Frete'].mean()
mean_frete_loja2 = loja2['Frete'].mean()
mean_frete_loja3 = loja3['Frete'].mean()
mean_frete_loja4 = loja4['Frete'].mean()

# Create a DataFrame for plotting
fretes_medios = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete Médio': [mean_frete_loja1, mean_frete_loja2, mean_frete_loja3, mean_frete_loja4]
})

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Loja', y='Frete Médio', data=fretes_medios, palette='mako')
plt.title('Média de Frete por Loja')
plt.xlabel('Loja')
plt.ylabel('Frete Médio (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

  A analise tem como objetivo, demonstrar através dos graficos e resultados qual loja do senhor Joao pode ser vendida para que ele possa aumentar seu capital e ampliar seu empreendimento.
  
  Para a analise, foram utilizados dados referentes a 4 lojas.
Com base nos resultados apresentados foi observado que a loja 4 apresenta faturamento inferior as demais lojas.
Quanto a vendas por categoria, a categoria eletrônicos se destaca como categoria mais vendida em todas as lojas, seguida de eletrodomésticos, móveis e instrumentos musicais. Dentre os produtos mais vendidos a loja 4 apresentou valores de vendas mais baixos comparados as demais lojas, mesmo quando analisãmos os produtos mais vendidos, como eletrônicos e eletrodomésticos.

Tendo em vista esse resultado, é possível concluir que a loja 4 não tem apresentado resultados satisfatórios e pode ser vendida para que o senhor João possa  ter capital sulficiente para expandir seus négócios.
